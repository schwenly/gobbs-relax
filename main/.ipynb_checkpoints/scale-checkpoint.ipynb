```python
Tnum = 1000
NNmax = 30.5
NNmin = 9.5
J = 0.3
Tmax = 40 * J


fsc = np.genfromtxt('sca.csv', delimiter=',')
psc = np.genfromtxt('ps123.csv', delimiter=',')

print(fsc.max())
```
